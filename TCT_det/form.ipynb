```python
map_uid = {}
for i in range(len(df_uid)):
    folder = df_uid['originPath'].iloc[i]
    uid = df_uid['seriesInstanceUID'].iloc[i]
    folder_name = folder.split('/')[0].split('_')[0]
    map_uid[folder_name] = uid
```

```python
map_type_ = {}
for i in range(len(df_ori)):
    folder = unicode(df_ori[u'扫描号'].iloc[i].split('_')[0])
    weishengwu = unicode(df_ori[u'微生物'].iloc[i])
    lingzhuang = unicode(df_ori[u'鳞状上皮细胞异常'].iloc[i])
    xianzhuang = unicode(df_ori[u'腺上皮细胞异常'].iloc[i])
    time = unicode(df_ori[u'数据采集时间'].iloc[i])
    source = unicode(df_ori[u'数据来源'].iloc[i])
    place = unicode(df_ori[u'地区'].iloc[i])
    province = unicode(df_ori[u'省份'].iloc[i])
    changj = unicode(df_ori[u'场景'].iloc[i])
    type_ = unicode(df_ori[u'阴阳性'].iloc[i]) 
    if i >= 2:
        df_model.loc[i] = [randint(-1,1) for _ in range(len(df_model.columns))]
    try:
        df_model.iloc[i,0] = map_uid[folder]
        map_type_[map_uid[folder]] = type_
    except:
        df_model.iloc[i,0] = u'bbed7d5c903a53d285201d4bc62b2752ac535f65'
        map_type_[u'bbed7d5c903a53d285201d4bc62b2752ac535f65'] = type_
    df_model.iloc[i,1] = 'PATHOLOGY'
    df_model.iloc[i,2] = None
    df_model.iloc[i,3] = 'WOOB'
    df_model.iloc[i,4] = None
    df_model.iloc[i,5] = u'{' + u'微生物: ' + weishengwu + u', 鳞状上皮: ' + lingzhuang + u', 腺上皮: ' + xianzhuang + u'}'
    df_model.iloc[i,6] = u'{VER:1}'
    df_model.iloc[i,7] =  u'{'  + u'数据采集时间: ' + time + u', 数据来源: ' + source + u', 地区: ' + place + u', 省份: ' + \
                        province + u', 场景: ' + place + u', 阴阳性: ' + type_ + u'}'
    df_model.iloc[i,8] = None
    df_model.iloc[i,9] = None
    
```

```python
for uid in map_type_:
    if map_type_[uid] == u'\u9634':
        print uid
```

```python
df_model.to_excel(writer, 'Sheet1')
writer.save()
```
